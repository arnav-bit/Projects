```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics
```

```python
raw_data = pd.read_csv('/Users/zaidfarooqui/Desktop/Studies/External_Courses/Python_Projects/Python_Files/Data-04-05-customer-lifetime-value.csv')
```

```python
print(raw_data.dtypes)
print(raw_data.head)
```

Cleaning the data to remove customer id which is not relevant 

```python
cleaned_data = raw_data.drop('CUST_ID', axis = 'columns')
```

Correlation of months to CLV

```python
print(cleaned_data.corr()['CLV'])
```

Creating the train test split, define predictors and targets

```python
predictors = cleaned_data.drop('CLV', axis = 'columns')
targets = cleaned_data.CLV
```

```python
pred_train, pred_test, tar_train, tar_test = train_test_split(predictors, targets, test_size= 0.1)
print('Predictor Training size is:', pred_train.shape)
```

Building the model

When building a linear regression model, you get 2 variables, coefficient and intercept. <br>
For y = mx + c, y is the target, x is the predictor, m is the regression coefficient (slope of the line) amd c is the intercept

```python
model = LinearRegression()
model.fit(pred_train.values, tar_train.values)
print("Model coefficient is:\n", model.coef_)
print("Model intercept is : \n", model.intercept_)
```

Testing on the data

```python
predictions = model.predict(pred_test.values)
print("Prediction is:", predictions)
```

Printing accuracy of data<br>
For checking accuracy or similar variables, we always use tar_test data

```python
r2_score = sklearn.metrics.r2_score(tar_test, predictions)
print("R2 score is: ",r2_score)
```
