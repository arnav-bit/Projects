```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import sklearn.metrics
from sklearn.naive_bayes import GaussianNB
```

```python
prospect_data = pd.read_csv('/Users/zaidfarooqui/Desktop/Studies/External_Courses/Python_Projects/Python_Files/Data-02-05-web-browsing-data.csv')
print(prospect_data.dtypes)
print(prospect_data.head())
```

Describe data - perform statistical analysis on the data across all columns

```python
print(prospect_data.describe())
```

Perform corelation of buy with other variables to understand which action has the highest corelation with buying

```python
print(prospect_data.corr()['BUY'])
```

Reduce the dataset to only columns having strong corelation with buy

Here we use predictors (to predict the value of the target) and target (whose value is modeled by the predictor)).

Set predictors to columns with high corelation and target to BUY

```python
predictors = prospect_data[['REVIEWS', 'BRO_TOGETHER', 'COMPARE_SIMILAR', 'WARRANTY',  'SPONSORED_LINKS']]
targets = prospect_data.BUY
```

Define your training testing model with 70-30 ratio, define pred_train and test 

```python
pred_train, pred_test, tar_train, tar_test = train_test_split(predictors, targets, test_size=.3)
```

Shape gives the dimension of your data. shape of 350,5 means your pred_train has 350 rows and 5 columns like we declared above

```python
print('Predictor Training Size', pred_train.shape, 'Predictor Testing Size', tar_train.shape)
```

Building models and checking accuracy. you always FIT the model with predictors training values and target training values

```python
classifier = GaussianNB()
classifier = classifier.fit(pred_train.values, tar_train.values)
```

After the model is created, you need to define predictions - what should be. this is the testing part of your predictors. after you have this, you can compare your training_test for accuray

```python
predictions = classifier.predict(pred_test.values)
```

```python
print(sklearn.metrics.accuracy_score(tar_test, predictions))
```

Doing a probability test for a new comer

```python
pred_prob = classifier.predict_proba(pred_test.values)
pred_prob = pred_prob[0,1]
print('The probability of purchase is :', pred_prob)
```
